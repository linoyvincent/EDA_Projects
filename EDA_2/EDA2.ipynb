### Introduction to Exploratory Data Analysis (EDA) for Supermarket Sales Dataset

The **Supermarket Sales** dataset contains **1,000 transactions** from a supermarket, covering various aspects of sales, customer details, and financial metrics. It consists of **17 columns**, including categorical and numerical variables.

Key columns:
- **Invoice ID**: Unique identifier for each transaction.
- **Branch & City**: Identifies the store location.
- **Customer type & Gender**: Provides demographic details.
- **Product line**: Category of the purchased product.
- **Unit price & Quantity**: Determines the cost of individual products and purchase amounts.

5. **Time-Series Analysis**: Identifying sales trends over time.

### **Checking outlier using boxplot**

```python
for i in df.select_dtypes(include='number').columns:
    sns.boxplot(data=df, x = i)
    plt.show()
```

### **Removing outliers using Interquartile Range**

```python
def remove_outliers_iqr(df, columns):
    for column in columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
        return df


df = remove_outliers_iqr(df, df.select_dtypes(include='number').columns)
```

## **Exploratory Data Analysis(EDA) Questions**

1. What is the total sales (Total) for each product line?

```python
plt.figure(figsize=(8, 6))
product_sales = df.groupby('Product_line')['Total'].sum().sort_values()
sns.barplot(x=product_sales.values, y=product_sales.index, palette='viridis')
plt.title('Total Sales by Product Line')
plt.xlabel('Total Sales')
plt.ylabel('Product Line')
plt.tight_layout()
plt.show()
```

2. How do daily sales (Total) trend over time?

```python
daily_sales = df.groupby(df['DateTime'].dt.date)['Total'].sum()
plt.figure(figsize=(10, 6))
sns.lineplot(x=daily_sales.index, y=daily_sales.values)
plt.title('Daily Sales Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

3. What is the distribution of product ratings?

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Rating'], bins=10, kde=True, color='skyblue')
plt.title('Distribution of Product Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

4. How does gross income (gross_income) vary across different product lines?

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Product_line', y='gross_income', data=df, palette='Set2')
plt.title('Gross Income Distribution by Product Line')
plt.xlabel('Product Line')
plt.ylabel('Gross Income')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

5. Is there a relationship between unit price and quantity sold?

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Unit_price', y='Quantity', data=df, hue='Product_line', palette='deep')
plt.title('Unit Price vs Quantity Sold')
plt.xlabel('Unit Price')
plt.ylabel('Quantity Sold')
plt.tight_layout()
plt.show()
```

6. What is the distribution of payment methods?

```python
payment_counts = df['Payment'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(payment_counts, labels=payment_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Payment Method Distribution')
plt.tight_layout()
plt.show()
```

## **Summary**

This Exploratory Data Analysis (EDA) of the supermarket sales dataset provides valuable insights into the store's operations, customer behavior, and sales performance. The dataset was first cleaned and prepared for analysis by standardizing column names and converting the Date and Time columns into a combined DateTime format. This allowed for more accurate time-based analysis. The data was checked for missing and duplicate values, and none were found, ensuring the dataset was complete and ready for in-depth exploration.

Descriptive statistics revealed that the numerical columns such as Unit_price, Quantity, Total, and gross_income had a wide range of values. To prevent extreme values from skewing the analysis, outliers were identified using box plots and removed using the Interquartile Range (IQR) method. This step improved the accuracy of the analysis by ensuring that unusually high or low values did not distort the data distribution.

The sales analysis highlighted that the Food and beverages and Electronic accessories product lines generated the highest total sales. A line chart of daily sales showed noticeable peaks, indicating certain days with exceptionally high sales, possibly due to promotions or seasonal demand. The scatter plot analysis between Unit_price and Quantity revealed a weak correlation, suggesting that product pricing did not significantly impact how much customers purchased. Additionally, a pie chart of payment methods indicated that E-wallets were the most popular mode of payment, slightly outperforming cash and credit card transactions.

Customer behavior analysis revealed that 'Member' customers made more purchases compared to 'Normal' customers, highlighting the effectiveness of the loyalty program. Sales were almost equally distributed between male and female customers. Branch-level analysis showed that 'Branch A' achieved the highest customer satisfaction based on average ratings, suggesting stronger operational efficiency or better customer service in that branch.

The correlation heatmap revealed strong positive relationships between 'Total', 'cogs', and 'gross_income', which is expected in sales data since these financial metrics are inherently linked. Grouped analysis by product line and branch confirmed that 'Food and beverages' consistently outperformed other categories, and 'Branch A' led in both sales and customer satisfaction.

Based on these insights, it is recommended that the supermarket expands its best-selling product lines, particularly 'Food and Beverages', and strengthens loyalty programs to convert more 'Normal' customers into 'Members'. Introducing targeted promotions during low-sales periods could help stabilize sales fluctuations. Additionally, leveraging the popularity of 'E-wallet' payments through exclusive discounts or cashback offers could further boost sales. Overall, the EDA provided a comprehensive understanding of the supermarket's operations and highlighted actionable strategies for improving performance.
