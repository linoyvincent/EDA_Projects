## **Exploratory Data Analysis(EDA) Questions**

1. What is the distribution of customers across different countries, states, and cities?

```python
country_distribution = df.groupby('Country')['CustomerID'].count()
print(country_distribution)
```

```python
plt.figure(figsize=(10, 6))
country_distribution.plot(kind='bar', color='red')
plt.title('Customer Distribution by Country')
plt.xlabel('Country')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45)
plt.show()
```

```python
state_distribution = df['State'].value_counts()
print(state_distribution)
```

```python
plt.figure(figsize=(10, 6))
state_distribution.plot(kind='bar', color='red')
plt.title('Customer Distribution by State')
plt.xlabel('State')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45)
plt.show()
```

```python
city_distribution = df['City'].value_counts()
print(city_distribution)
```

```python
plt.figure(figsize=(10, 6))
city_distribution.plot(kind='bar', color='red')
plt.title('Customer Distribution by City')
plt.xlabel('City')
plt.ylabel('Number of Customers')
plt.xticks(rotation=90)
plt.show()
```

2. What are the age and gender distributions of the customers?

```python
plt.figure(figsize=(10, 6))
df['Age'].plot(kind='hist', bins=20, color='blue')
plt.title('Age Distribution of Customers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
df['Gender'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['pink', 'blue'])
plt.title('Gender Distribution of Customers')
plt.ylabel('')
plt.show()
```

3. How does monthly income vary across different countries?

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Country', y='MonthlyIncome', data=df)
plt.title('Income Distribution by Country')
plt.xticks(rotation=45)
plt.show()
```

Plotting Income vs Credit Score Distribution by Gender

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='MonthlyIncome', y='CreditScore', data=df, hue='Gender')
plt.title('Income vs Credit Score by Gender')
plt.xlabel('Monthly Income')
plt.ylabel('Credit Score')
plt.show()
```

4. What is the average revenue generated from different age?

```python
df['Revenue'] = df['Price'] * df['Quantity']
```

```python
average_revenue = df.groupby('Age')['Revenue'].mean()
plt.figure(figsize=(10, 6))
average_revenue.plot(kind='bar', color='orange')
plt.title('Average Revenue by Age')
plt.xlabel('Age')
plt.ylabel('Average Revenue')
plt.show()
```

5. What is the average revenue generated by campaign schema?

```python
campaign_performance = df.groupby('CampaignSchema').agg({'Revenue': 'mean'})
print(campaign_performance)
```

6. How does revenue change over time?

```python
df['YearMonth'] = df['OrderConfirmationTime'].dt.to_period('M')
```

```python
monthly_revenue = df.groupby('YearMonth')['Revenue'].sum()
print(monthly_revenue)
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x=monthly_revenue.index.astype(str), y=monthly_revenue.values, marker='o', color='purple')
plt.title('Monthly Revenue Trend', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Revenue', fontsize=12)
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

7. How many customers are there below 40 years old?

```python
customers = df.loc[(df['Age'] <= 40)].shape[0]
print(f"Number of customers below age 40: {customers}")
```

8. How many confirmed orders are there from 'Instagram Ads'?

```python
confirmed_orders_instagram = df.loc[(df['CampaignSchema'] == 'Instagram-ads') & (df['OrderConfirmation'] == True)]
num_confirmed_orders_instagram = confirmed_orders_instagram.shape[0]
print(f"Number of confirmed orders from Instagram Ads: {num_confirmed_orders_instagram}")
```

9. What is the total quantity of products sold to customers gender wise?

```python
male_customers_quantity = df.loc[df['Gender'] == 'Male', 'Quantity'].sum()
print(f"Total quantity of products sold to male customers: {male_customers_quantity}")

female_customers_quantity = df.loc[df['Gender'] == 'Female', 'Quantity'].sum()
print(f"Total quantity of products sold to female customers: {female_customers_quantity}")
```

10. What is the average income for customers in different countries?

```python
average_income_by_country = df.groupby('Country').apply(lambda x: x.loc[:, 'MonthlyIncome'].mean())
print(average_income_by_country)
```

```python
plt.figure(figsize=(10, 6))
average_income_by_country.plot(kind='bar', color='blue')
plt.title('Average Income of Customers by Country')
plt.xlabel('Country')
plt.ylabel('Average Income')
plt.show()
```

11. What is the total revenue generated from each country?

```python
country_revenue = df.groupby('Country')['Revenue'].sum()
print(country_revenue)
```

```python
plt.figure(figsize=(10, 6))
country_revenue.plot(kind='bar', color='blue')
plt.title('Revenue Generated from each Country')
plt.xlabel('Country')
plt.ylabel('Revenue Generated')
plt.show()
```

12. How many products are sold for different category?

```python
category_sold = df.groupby('Category')['Quantity'].sum()
print(category_sold)
```

```python
plt.figure(figsize=(10, 6))
category_sold.plot(kind='bar', color='green')
plt.title('Different Categories of Products Sold')
plt.xlabel('Different Categories')
plt.ylabel('No: of Products Sold')
plt.show()
```

13. What are the different payment methods used for transaction?

```python
pay_method = df.groupby('PaymentMethod')['CustomerID'].count()
print(pay_method)
```

```python
plt.figure(figsize=(8, 8))
pay_method.plot(kind='pie', autopct='%1.1f%%')
plt.title('Payment Methods Used for Transaction')
plt.ylabel('')
plt.show()
```

14. what percentage of orders were returned?

```python
returned = df.loc[df['OrderReturn'] == True].shape[0]
not_returned = df.loc[df['OrderReturn'] == False].shape[0]
total_orders = returned + not_returned
percent = (returned / total_orders) * 100
print(f'No: of orders returned is {returned}')
print(f'Percentage of orders returned is {percent}')
```

15. What is the average credit score for each gender?

```python
average_credit_score_gender = df.groupby('Gender')['CreditScore'].mean()
print(average_credit_score_gender)
```

16. What is the total quantity sold for each product?

```python
category_quantity_sold = df.groupby('Product')['Quantity'].sum()
print(category_quantity_sold)
```

17. What is the average income for customers in different age group?

```python
bins = [0, 30, 40, 50, 60, 100]
labels = ['18-30', '31-40', '41-50', '51-60', '60+']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

average_income_agegroup = df.groupby('AgeGroup')['MonthlyIncome'].mean()
print(average_income_agegroup)
```

```python
plt.figure(figsize=(10, 6))
average_income_agegroup.plot(kind='bar', color='green')
plt.title('Average Income for Customers in Different Age Groups')
plt.xlabel('Different Age Groups')
plt.ylabel('Average Income')
plt.show()
```

18. How many orders did customers place in each month?

```python
monthly_orders = df.groupby('YearMonth')['CustomerID'].count()
print(monthly_orders)
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x=monthly_orders.index.astype(str), y=monthly_orders.values, marker='o', color='purple')
plt.title('Monthly Orders Trend', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Orders', fontsize=12)
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

19. How many items were ordered by different Campaign Schema?

```python
items_campaign = df.groupby('CampaignSchema')['Quantity'].count()
print(items_campaign)
```

```python
plt.figure(figsize=(10, 6))
items_campaign.plot(kind='bar', color='green')
plt.title('Items Ordered by Different Campaign Schema')
plt.xlabel('Different Campaign Schema')
plt.ylabel('Items Ordered')
plt.show()
```

This Exploratory Data Analysis (EDA) project centers on analyzing customer online shopping data to uncover patterns, relationships, and actionable insights. Descriptive statistics summarized the central tendencies and variability of numerical columns, while a correlation heatmap provided a clear visual representation of inter-variable relationships.

The analysis uncovered significant insights into customer behavior and financial patterns. Bar charts illustrated the geographic distribution of customers across countries, states, and cities, while histograms and pie charts showcased age and gender distributions. Financial patterns, such as variations in monthly income across countries and trends in income versus credit score by gender, were highlighted using box plots and scatter plots.
